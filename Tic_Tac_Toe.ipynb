```python
moves = {'1': ' ', '2': ' ', '3': ' ', '4': ' ', '5': ' ', '6': ' ', '7': ' ' , '8': ' ', '9': ' '}


def board(moves):
    
    v20 = '      |      |      '
    h20 = '--------------------'
    t1 = '  {}   |   {}  |   {}   '.format(moves['1'],moves['2'],moves['3'])
    t2 = '  {}   |   {}  |   {}   '.format(moves['4'],moves['5'],moves['6'])
    t3 = '  {}   |   {}  |   {}   '.format(moves['7'],moves['8'],moves['9'])
    
    print(v20)
    print(t1)
    print(v20)
    print(h20)
    
    print(v20)
    print(t2)
    print(v20)
    
    print(h20)
    print(v20)
    print(t3)
    print(v20)
```

```python
def assign():
    
    players = {'p1': 'assign', 'p2': 'assign'}
    choice = input('Player 1: Please choose "X" or "O": ').upper()
    
    
    while choice not in ['X', 'O']:
        choice = input('Player 1: Invalid input. Please choose "X" or "O": ').upper()
        
        
    players['p1'] = choice
    
    
    if players['p1'] == 'X':
        players['p2'] = 'O'
    
    else: 
        players['p2'] = 'X'
    
    
    print('Player 1 chooses: ' + players['p1'] + '\nPlayer 2 is autoset to: ' + players['p2'])
    
    return players
    
```

```python
def game(assign, moves):

    players = assign
    counter = 1
    
    while counter != 10: 
        
        if counter % 2 != 0:
            mark = input('Player 1: Place your mark on the board from 1-9: ')
                
            while extent(int(mark)) == False:
                mark = input('Player 1: Out of range. Place your mark on the board from 1-9: ')
                
            while empty(moves, mark) == False:
                mark = input('Player 1: Spot has been taken. Please choose again. ')
                    
                if extent(int(mark)) == False:
                    mark = input('Player 1: Out or range. Please choose from 1-9: ')   
                    
            if empty(moves, mark):
                moves[mark] = players['p1']
                counter += 1
                board(moves)
                
                if win(moves):
                    print('Player 1 wins!!')
                    break
                              
        elif counter % 2 == 0:
            mark = input('Player 2: Place your mark on the board from 1-9: ')
            
            while extent(int(mark)) == False:
                mark = input('Player 2: Out of range. Place your mark on the board from 1-9: ')
            
            while empty(moves, mark) == False:
                mark = input('Player 2: Spot has been taken. Please choose again. ')
                
                if extent(int(mark)) == False:
                    mark = input('Player 2: Out or range. Please choose from 1-9: ')
                    
            if empty(moves, mark):
                moves[mark] = players['p2']
                counter += 1
                board(moves)
                
                if win(moves):
                    print('Player 2 wins!!')
                    break
```

```python
def extent(mark):
    
    if mark in list(range(1,10)):
        return True
    
    else:
        return False
```

```python
def empty(moves, mark):
    
    if moves[mark] == 'X' or moves[mark] == 'O':
        return False
    
    else:
        return True    
```

```python
def win(moves):
    
    m = moves
    w_combos = {'v1': [m['1'],m['4'],m['7']], 'v2': [m['2'],m['5'],m['8']], 'v3': [m['3'],m['6'],m['9']], 
                'h1': [m['1'],m['2'],m['3']], 'h2': [m['4'],m['5'],m['6']], 'h3': [m['7'],m['8'],m['9']], 
                'c1': [m['1'],m['5'],m['9']], 'c2': [m['3'],m['5'],m['7']]}
    
    for k,v in w_combos.items():
        
        if ' ' in v :
            continue
            
        elif v[0] == v[1] and v[1] == v[2] and v[0] == v[2]:
            print('Winning section at: '+ k)
            return True
```

```python
def again():
    

    choice = input('Do you want to play again? Y or N: ').upper()
    
    while choice not in ['Y', 'N']:
        choice = input('Player 1: Invalid input. Please choose "Y" or "N": ').upper()
    
    if choice == 'Y':
        global moves
        moves = {'1': ' ', '2': ' ', '3': ' ', '4': ' ', '5': ' ', '6': ' ', '7': ' ' , '8': ' ', '9': ' '}
        
        return start()
    
    else:
        print('Good game!') 
```

```python
def start():
    board(moves)
    game(assign(), moves)
    again()

start()
```
